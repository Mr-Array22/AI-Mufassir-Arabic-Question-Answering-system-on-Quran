```python
import pandas as pd
```

```python
column = ["qid", "Q0", "docno", "rank", "score", "tag"]
df1 = pd.read_table('bigIR_run02.tsv', names=column)



df2 = pd.read_table('bigIR_run04_ArabicBERT.tsv', names=column)
```

```python
df1
```

```python
df2
```

```python
qid2=[]
Q02=[]
docno2=[]
score2=[]
tag2=[]
rank2=[]

flage=df1["qid"][0]
num=1
listate=[]
for index, row in df1.iterrows():
  if flage != row["qid"]:
    num=1
    flage = row["qid"]
    qid2.append(row["qid"])
    Q02.append(row["Q0"])
    docno2.append(row["docno"])
    score2.append(row["score"])
    tag2.append(row["tag"])
    rank2.append(num)
  else:
    if num <= 10:
      qid2.append(row["qid"])
      Q02.append(row["Q0"])
      docno2.append(row["docno"])
      score2.append(row["score"])
      tag2.append(row["tag"])
      rank2.append(num)

  num=num+1

dataa2 = {"qid":qid2, "Q0":Q02, "docno":docno2, "rank":rank2, "score":score2, "tag":tag2}


df1_final = pd.DataFrame(dataa2)
df1_final
```

```python
df1_final['tag'] = 'bigIR_run03'
```

```python
df1_final
```

```python
df1_final.to_csv("bigIR_run03.tsv", sep='\t', index=False, header=False)
```
