#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scores=pd.read_json("scores.json").T
```

```python
scores.head()
```

```python
scores=scores.reset_index()
scores=scores.rename(columns={'index': 'model'})
```

```python
scores.head()
```

```python
scores['model']=scores['model'].str.split('/').str[1]
```

```python
scores
```

```python
scores['model'].values
```

```python
scores.columns
```

```python
scores.plot.barh(x='model', y=[ 'pRR', 'exact_match', 'f1'], rot=0)

plt.title('Scores by model')
plt.xlabel('Scores')
plt.ylabel('Model')

plt.show()
```
